## Standard Deviation + Varience 
Are measures of spread of data  

```python
# Varience = The measure of the average degree to which each number is different from the main
# Higher varience = wider range of numbers
# Lower varience = lower range of numbers

high_var_array = np.array([1, 100, 200, 300, 4000, 5000])
low_var_array = np.array([2, 4, 6, 8, 10])
```

```python
np.var(high_var_array), np.var(low_var_array)
```

```python
# Standard Deviation = The measure of how spread out a group of numbers are from the mean

np.std(high_var_array), np.std(low_var_array)
```

```python
np.mean(high_var_array), np.mean(low_var_array)
```

```python
import matplotlib.pyplot as plt

plt.hist(high_var_array)
plt.show()
```

```python
plt.hist(low_var_array)
plt.show()
```
