```python
fav_food_prices = {"Tacos": 14,
                   "Burger": 12,
                   "Sushi": 30}
```

```python
fig, ax = plt.subplots()
ax.bar(fav_food_prices.keys(), fav_food_prices.values());
ax.set(title="My Favorite Foods",
      xlabel="Favorite Foods",
      ylabel="Price");
```

```python
fig, ax = plt.subplots()
ax.barh(fav_food_prices.keys(), fav_food_prices.values())
ax.set(title="My Favorite Foods",
      xlabel="Prices",
      ylabel="Favorite Foods");
```
