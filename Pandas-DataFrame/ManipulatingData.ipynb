```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
```

# Manipulating Data

### Upper & Lower case

```python
# Lower case letters
car_sales["Make"].str.lower()
```

```python
# Upper case letters 
car_sales["Make"].str.upper()
```

```python
car_sales
```

```python
# These changes wont automatically save to your df
# youll have to reassign that column
car_sales["Make"] = car_sales["Make"].str.upper()
```

```python
car_sales
```

### Filling Missing Data

```python
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
# Changes wont be made to the data
car_sales_missing
```

```python
# setting missing data in the df
# By using inplace. When set to True = fill in place this will modify other views on object
# You could also just use the reassignment method

car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(),
                                    inplace=True)
car_sales_missing
```

```python
car_sales_missing.dropna()
```

```python
# It works but we wont be able to get this data back and would have to reimport the data again
car_sales_missing.dropna(inplace=True)
car_sales_missing
```

```python
# Assigning to limit lost data
# Reset Dataframe
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
# Returns a copy of car_sales_missing butt with dropped values
car_sales_missing_dropped = car_sales_missing.dropna()
car_sales_missing_dropped
```

```python
car_sales_missing_dropped.to_csv("car-sales-missing-dropped.csv")
```
