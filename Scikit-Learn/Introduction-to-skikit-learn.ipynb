```python
import numpy as np
```

```python
# Get the data ready
import pandas as pd
heart_disease = pd.read_csv("csv-files/heart-disease.csv")
heart_disease
```

```python
# Create x (features matrix)
x = heart_disease.drop("target", axis=1) 

# Create your own (labels)
y = heart_disease["target"]
```

```python
import warnings 
warnings.filterwarnings("default")
```

```python
# 2. Choose the right model and hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# We'll keep the default hyperparameters
clf.get_params()
```

```python
# 3. Fit the model to the training data 
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
import sklearn
sklearn.show_versions()
```

```python
clf.fit(x_train, y_train);
```

```python
x_train
```

```python
x_test
```

```python
y_preds = clf.predict(x_test)
y_preds
```

```python
y_test
```

```python
# 4. Evaluate the model (How good are the predictions) on the training data and the test data
clf.score(x_train, y_train)

```

```python
clf.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# 5. Imporve the model 
# Try different amount of n_estimators
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
    print(f"Model accuracy on test set: {clf.score(x_test, y_test) * 100:.2f}%")
    print(" ")
```

```python
# 6. Save a model and load it (wb = write binary rb= read binary )
import pickle 

pickle.dump(clf, open("random_forst_model_1.pk1", "wb"))
```

```python
loaded_model = pickle.load(open("random_forst_model_1.pk1", "rb"))
loaded_model.score(x_test, y_test)
```
